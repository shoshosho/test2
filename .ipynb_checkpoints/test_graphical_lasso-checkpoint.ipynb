```python
import pandas as pd
import seaborn as sns
dt = pd.read_csv("/Users/01017387/PycharmProjects/test/uploads/2015_hoyu_daisu.csv")
tgtvar = "乗用車"
```

```python
#graph lassoのテスト
#→L1を噛ませた逆相関行列

from sklearn.covariance import GraphLasso
# class sklearn.covariance.GraphLasso(alpha=0.01, mode='cd', tol=0.0001, enet_tol=0.0001, max_iter=100, verbose=False, assume_centered=False)
from sklearn.covariance import GraphLassoCV
# class sklearn.covariance.GraphLassoCV(alphas=4, n_refinements=4, cv=None, tol=0.0001, enet_tol=0.0001, max_iter=100, mode='cd', n_jobs=1, verbose=False, assume_centered=False)
#標準化が必要なので、標準化用のimportも必要
from sklearn.preprocessing import StandardScaler
# class sklearn.preprocessing.StandardScaler(copy=True, with_mean=True, with_std=True)


```

```python
st = StandardScaler()
y = dt["乗用車"]
d = dt.drop(["都道府県名","乗用車"],axis=1)
dd = st.fit_transform(d)
#ddはnpになってる
```

```python
gl = GraphLasso(alpha=0.1)
glcv = GraphLassoCV(cv=4,n_jobs=4)
gl.fit(dd)
glcv.fit(dd)
```

```python
import numpy as np
from numpy.linalg import inv
todf = gl.get_precision()
todf_inv = inv(todf)
df = pd.DataFrame(todf,columns=d.columns,index=d.columns)
dfinv = pd.DataFrame(todf_inv,columns=d.columns,index=d.columns)
df
```

```python
target = dt[tgtvar]
dt_for_model = dt.ix[:,["普通","軽"]]
# dt_for_model = dt.ix[:,var_lis]
#連続値だった場合
```

```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error

if len(target.unique()) != 2:
    #線形回帰
    # reg = linear_model.LinearRegression()
    reg = linear_model.Lasso(alpha=0.5)
    reg.fit(dt_for_model.values,target.values)
    coefs = reg.coef_
    reg_res = reg.predict(dt_for_model.values)
    reg_score = mean_squared_error(target.values, reg_res)
```
