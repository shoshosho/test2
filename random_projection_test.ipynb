```python
import pandas as pd
import numpy as np

from sklearn.random_projection import johnson_lindenstrauss_min_dim
from sklearn import random_projection
```

```python
#データ
data = pd.read_csv("/Users/01017387/Desktop/しごと/data/car用/加工済み/2015_hoyu_daisu.csv",dtype="str")
data.head()
```

```python
#余計な文字を取り除く
data = data.drop("都道府県名",axis=1).applymap(lambda x : x.replace(",","").replace("%","")).astype(float)
data.head()
```

```python
X = data.drop("乗用車",axis=1).values
y = data["乗用車"].values
```

```python
#random projection
#3つほどパターンがあって、それぞれ多分違った特徴が出る
# 1.The Johnson-Lindenstrauss lemma
# →次元数がどのくらいまでに減るかを可視化できる
# 2.Gaussian random projection
#基本的には乱数で生成した遷移行列を元に、違う次元へ写像する（変数間のユークリッド距離を保ったまま）というのがアイデアらしい
#基本的には100変数以上でないと使う意味がない模様（そもそもエラーが返ってくる）
transformer = random_projection.GaussianRandomProjection(eps=0.1)
X_new = transformer.fit_transform(X)
X_new.shape
```

```python
# 3.Sparse random projection
transformer = random_projection.SparseRandomProjection()
X_new = transformer.fit_transform(X)
X_new.shape
```
