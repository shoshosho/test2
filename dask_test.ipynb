```python
import numpy as np
import pandas as pd
import dask.dataframe as dd
import dask.threaded
from dask import delayed, compute
import dask.multiprocessing
import dask.multithreading
```

```python
# list(range(1000))
len(np.arange(1000,20))
```

```python
df = pd.DataFrame({'X': np.arange(100000000), 
                   'Y': np.arange(100000000),
                   'Z': np.arange(100000000)},
                  index=list(range(100000000)))
```

```python
ddf = dd.from_pandas(df, 10)
```

```python
print(ddf.columns)
# ('X', 'Y', 'Z')

# DataFrame の index
print(ddf.index)
# dd.Index<from_pandas-b08addf72f0693a9fa1bb6c21d91f3d4-index, divisions=('a', 'f', 'j')>

# DataFrame の divisions (パーティションの分割箇所)
print(ddf.divisions)
# ('a', 'f', 'j')
 
# DataFrame のパーティション数
print(ddf.npartitions)
```

```python
ddf.cumsum().compute()
```

```python
def foo(arg):
    return arg.cumsum().compute()

jobs = [delayed(foo)(ddf)]

compute(*jobs, get=dask.multiprocessing.get)
# compute(*jobs, get=dask.multithreading.get)
#オーバーヘッドが大きそう、全然プロセスが始まらない
```
